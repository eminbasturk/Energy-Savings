## Geri dönüşümden tasarruf edilen enerji
<p>Geri dönüşümün, malzemeleri sıfırdan yapma ihtiyacını azaltarak veya ortadan kaldırarak enerji tasarrufu sağladığını biliyor muydunuz? Örneğin, alüminyum kutu üreticileri, geri dönüştürülmüş kutuları temizleyerek ve eriterek cevherden alüminyum üretmenin enerji maliyetli sürecini atlayabilir. Alüminyum, demir içermeyen bir metal olarak sınıflandırılır.</p>

<p>Singapur, sıfır atık ülkesi olma konusunda iddialı bir hedefe sahip. Singapur'da bertaraf edilen atık miktarı son 40 yılda yedi kat arttı. Bu hızda, Singapur’un tek katı atık sahası olan Semakau Katı Atık Sahası 2035 yılına kadar tükenecek. Sorunları daha da kötüleştiren Singapur, yeni yakma tesisleri veya atık depolama alanları inşa etmek için sınırlı alana sahip.</p>

<p>Hükümet, birleşik geri dönüşüm çabalarının her yıl kurtardığı toplam enerjiyi paylaşarak vatandaşları motive etmek istiyor. Onlara yardım etmeni istediler.</p>

### Singapur, 2015 ile 2019 arasında cam, plastik, demir ve demir dışı metalleri geri dönüştürerek kiloWatt saat (kWh) cinsinden yılda ne kadar enerji tasarrufu yaptı?

```python
import pandas as pd
waste_15_16_17 = pd.read_csv("datasets/wastestats.csv")
waste_15_16_17 = waste_15_16_17[["waste_type", "total_waste_recycled_tonne", "year"]]
waste_15_16_17 = waste_15_16_17[waste_15_16_17["waste_type"].isin(["Plastics", "Ferrous metal", "Non-ferrous metal", "Glass", "Ferrous Metal", 'Non-ferrous Metals', 'Ferrous Metals', 'Non-ferrous metals', "Non-ferrous metals", "Plastic"])]
waste_15_16_17.query("year >= 2015", inplace=True)
waste_15_16_17["total_waste_recycled_tonne"] = [int(each) for each in waste_15_16_17["total_waste_recycled_tonne"]]
waste_15_16_17["waste_type"] = ["Plastics" if each == "Plastic" else each for each in waste_15_16_17.waste_type]
waste_15_16_17["waste_type"] = ["Ferrous Metal" if each == "Ferrous metal" else each for each in waste_15_16_17.waste_type]
waste_15_16_17["waste_type"] = ["Non-Ferrous Metal" if each == "Non-ferrous metal" else ("Non-Ferrous Metal" if each == "Non-ferrous metals" else each) for each in waste_15_16_17.waste_type]
waste_15_16_17 = waste_15_16_17.sort_values("year", ascending = False)
waste_15_16_17 = waste_15_16_17.reset_index(drop = True)
waste_15_16_17 = waste_15_16_17.rename(columns = {"total_waste_recycled_tonne":"total_energy_saved"})
waste_15_16_17.head()
```
